Немного про типы данных в Python \
**int** целочисленный тип данных\
Происходит от слова **integer**, то есть целое число. \
Можно складывать, умножать, делить с остатком и находить остаток от деления, возводить в степень и тп.


```python
a = int(input())
print(type(a))

a = 57
b = 150
c = 10000000
print(type(a), type(b), type(c)) # Печатаем тип того что лежит в переменной функцией type()
print(a + b)
print(b - a)
print(b * 2)
print(5 ** 2, '= 5 ** 2') # Две звездочки возведение в степень
print(b // 100, '--->150 // 100 = 1, поскольку 100 помещается 1 раз') # Целочисленное деление, то есть сколько раз одно число полностью уместилось в другое
print(a % 2, '--->57 % 2 = 1, поскольку 57 дает остаток 1 при делении на 2') # Деление с остатком
print('\n') # обратный слеш обозначает, что идет специальный символ, например \n это переход на новую строчку, но это скорее бонус, можете пропускать)
a = int(input("Введите целое число номер 1:"))
b = int(input("Введите целое число номер 2:"))
print("Сумма ваших чисел -->", a + b)
print("Произведение ваших чисел -->", a * b)
print("Первое число в степени второго это -->", a ** b)
```

Второй тип данных строка **str** \
От английского **string** - строка \
Можно выводить, вводить, разбивать по разделителю, искать подслова и  
подстроки и тп, можно складывать и умножать на число.

```python
# вводят строку
# вывести (строку + ..) 3 раза подряд

# gg --> gg..gg..gg..

k = input()
s = "50"
print(s)
s2 = "60"
print(s2)
s3 = s + s2 # Сложили две строки, это называется конкатенация, когда вы приписываете первую строку ко второй
print(s3, '---> Сумма наших строчек')
print(s3 * 4, '---> Умножили строчку на 4 = Четыре раза написать ее подряд')
```

Третий тип данных это нецелое число, например $0.67$ \
Называется **float** - это переводится как плавать, казалось бы где связь? \
Дело в том, что такие числа еще называют числа с плавающей запятой или точкой отсюда **float**.\
 Можно делать все что и с обычными числами, только пропадает
деление с остатком и остаток.

```python
a = 10.0 # Обязательно обьявление через точку
b = 2.5
print(type(a), type(b))
print(a * b)
print(a ** b, '= 10 в степени 2.5') # Напоминаю, возведение в степень (показатель может быть нецелым)
print(a ** (0.5), '= корень из 10')
print('type of (0.5 * 2) -->', type(0.5 * 2), '--> Хотя произведение чисел целое, и даже одно из чисел целое, результат все равно имеет тип float')
```
